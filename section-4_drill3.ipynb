```python
from sklearn.datasets import load_wine

import pandas as pd  # fill in with the proper package name 
import numpy as np
```

```python
# Converts the load_wine() data from array to dataframe
def get_wine_data():
    data = load_wine() # contains data (the features and labels), the names for the columns
    features = data.data # equivalent to data["data"] (data is a dictionary)
    labels = data.target
    df = pd.DataFrame(features) # making a dataframe from a 2d array
    df.columns = data.feature_names # setting the names of the columns
    df["labels"] = labels
    return df
```

```python
# Get the names of the columns
def get_feature_titles():
    df = get_wine_data()
    return df.columns
```

```python
def add_2_to_magnesium():
    df = get_wine_data()
    df['magnesium'] += 2
    return df

add_2_to_magnesium()
```

```python
def create_flav_to_non_flav_ratio():
    df = get_wine_data()
    df["flav_to_non_flav_ratio"] = df["flavanoids"] / df["nonflavanoid_phenols"]
    return df.head()

df_head = create_flav_to_non_flav_ratio()
df_head
```

```python
def get_num_missing_values():
    df = get_wine_data()
    return df.isna().sum()
```

```python
def alcohol_content_level_sifter():
    df = get_wine_data()
    data_num = len(df.index)
    df["alcohol_content_level"] = 0
    for i in range(data_num):
        if df.loc[i, "alcohol"] >= 13.5:
            df.loc[i, "alcohol_content_level"] = 1
    return df

def alcohol_content_level_sifter2():
    df = get_wine_data()
    df1 = df[df["alcohol"] >= 13.5].copy()
    df2 = df[df["alcohol"] < 13.5].copy()
    df1["alcohol_content_level"] = 1
    df2["alcohol_content_level"] = 0
    df = pd.concat([df1, df2])
    df.sort_index(inplace=True)
    return df
```
